- Obtain the 50th Fibonacci number

        .data
        fib1 dw 0
        fib2 dw 1
        fib_num dw ?

        .code
        main proc
            mov ax, @data
            mov ds, ax

            mov cx, 49   
            mov bx, 0   

            ; Calculate Fibonacci sequence
            fib_loop:
                mov ax, fib2  
                add ax, fib1   
                mov bx, fib2   
                mov fib2, ax   
                mov fib1, bx   
                loop fib_loop

            mov fib_num, ax   

            ; Display the result
            mov ah, 09h
            mov dx, offset fib_msg
            int 21h

            mov ax, fib_num
            call display_number

            mov ah, 4Ch
            int 21h

        display_number proc
            push ax
            push bx
            push cx
            push dx

            xor bx, bx      ; Initialize BX to 0
            mov cx, 10      ; Set the divisor to 10
            mov dx, 0       ; Clear DX for the remainder

            display_loop:
                xor dx, dx  ; Clear DX before dividing
                div cx      ; Divide AX by CX
                add dl, '0' ; Convert the remainder to ASCII
                push dx     ; Push the ASCII character onto the stack
                inc bx      ; Increment the character count
                test ax, ax ; Check if AX is zero
                jnz display_loop

            ; Display the characters in reverse order
            pop dx
            mov ah, 02h
            int 21h

            loop display_loop

            pop dx
            pop cx
            pop bx
            pop ax

            ret
        display_number endp

        fib_msg db "The 50th Fibonacci number is: $"

        end main

### 4. What are the differences among sequential access, direct access, and random access?

Sequential access:
This is one of the simplest methods of accessing information, in this case the information is accessed sequentially, that is, the information is accessed in an ordered manner, where the first element would be accessed, then the second, then the third, and so on until the last element was reached.

Direct access:
In this method we can jump directly to a block or a certain information location without having to go through the previous blocks, where you can access element 5 without having to go through the previous elements.

Random access:
For random access we can access any memory location immediately and there is no time difference to be able to access different memory locations as in sequential or direct access, you you can access element 5 without having to go through the previous elements and without having to wait for the previous elements to be accessed.

### 5. What common characteristics are shared by all RAID levels?

Disk management: In all RAID levels, software or hardware is required to help manage how data is distributed among the disks and how information is read when it is required.

Increased performance: Most RAIDs seek to improve the performance of hard drives, distributing data across multiple hard drives, this allows data to be accessed more quickly and in turn improves writing.

Use of multiple disks: All RAID levels occupy more than one hard disk, this is something fundamental to the concept of it since more than one disk is needed to implement the RAID and also allows redundancy and increase of performance.

Hot Swap: In many RAID systems a hard drive can be hot swapped while the system is running, this allows a failing hard drive to be swapped out without powering down the system.

Redundancy: All RAID levels have redundancy, this is fundamental to the concept of RAID since it allows that if a hard drive fails, the information from the other hard drives can be recovered.
